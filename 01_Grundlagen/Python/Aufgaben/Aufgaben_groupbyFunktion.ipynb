# Anwendung der .groupby() Funktion in Python

Dieses Notebook dient der Anwendung der groupby Funktion innerhalb verschiedener Datensätze. Es wird stetig erweitert, um das Wissen und Neuerungen aktuell zu halten. Die Aufgaben sind in leicht, mittel und schwer aufgeteilt und beziehen sich auf das Dokument [car_insurance_claims.csv](https://github.com/EikeSuemper/Meine-Lernreise-Aktuariat-QuantFinance/blob/main/04_Datens%C3%A4tze/car_insurance_claims.csv)

```python
import numpy as np
import pandas as pd

df1 = pd.read_csv("https://raw.githubusercontent.com/EikeSuemper/Meine-Lernreise-Aktuariat-QuantFinance/refs/heads/main/04_Datens%C3%A4tze/car_insurance_claims.csv")
df1.head()
```

## Leichte Aufgaben

### Aufgabe 1

Gruppiere den DataFrame nach policy_state und berechne die Anzahl der Schadensfälle pro Bundesstaat.

```python
df1.groupby("policy_state").size()
```

```python
ergebnis1 = df1.groupby("policy_state").size().reset_index(name="Anzahl_Faelle")
ergebnis1.sort_values('Anzahl_Faelle', ascending=False)
```

Mit .size() werden die Anzahl der Zeilen je Staat gezählt und mit reset_index() die Serie in ein DataFrame gewandelt und mit der Spalte "Anzahl Fälle" aufgefüllt.

Es gibt 352 Schadenfälle in Ohio, 338 Schadenfälle in Illinois und 310 Schadenfälle in Indiana.

### Aufgabe 2

Gruppiere nach incident_severity und zeige die durchschnittliche Schadenssumme (total_claim_amount) für jeden Schweregrad.

```python
ergebnis2 = df1.groupby("incident_severity")["total_claim_amount"].mean().round(2)
ergebnis2.reset_index(name="Durchschnitt_Schaden_EUR")
```

### Aufgabe 3

Gruppiere nach insured_sex und berechne das durchschnittliche Alter (age) der Versicherten.

```python
ergebnis3 = df1.groupby('insured_sex')['age'].mean().round(2).reset_index(name='Durchschnitt_Alter')
ergebnis3
```

### Aufgabe 4

Gruppiere nach auto_make (Automarke) und zähle, wie viele Schadensfälle es pro Marke gibt.

```python
ergebnis4 = df1.groupby("auto_make").size().reset_index(name="anzahl_schadensfaelle")
ergebnis4
```

### Aufgabe 5

Gruppiere nach insured_sex und berechne den Mittelwert von umbrella_limit.

```python
ergebnis5 = df1.groupby("insured_sex")["umbrella_limit"].mean().reset_index(name="durchschnitt_umbrella_limit")
ergebnis5
```
