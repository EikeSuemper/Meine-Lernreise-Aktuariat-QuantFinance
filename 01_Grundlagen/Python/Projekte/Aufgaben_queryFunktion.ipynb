# Anwendung der .query() Funktion in Python

Dieses Notebook soll verschiedene Anwendungen der .query() aufzeigen und dessen Bedeutung darstellen. 

```python
import pandas as pd
import numpy as np

df = pd.read_csv("https://raw.githubusercontent.com/EikeSuemper/Meine-Lernreise-Aktuariat-QuantFinance/refs/heads/main/04_Datens%C3%A4tze/sales_data.xls")
df
```

### Aufgabe 1

Finde alle Bestellungen mit einem Gesamtumsatz (TotalSales) größer als 5000€.

```python
A = df.query("TotalSales >= 5000")
A[["OrderID","Product", "Quantity", "Date", "TotalSales"]].head()
```

### Aufgabe 2

Zeige alle Verkäufe in der Region 'North' oder 'South', bei denen die Menge (Quantity) mindestens 5 beträgt.

```python
B = df.query("Quantity >= 5 and Region in ['North', 'South']")
B[["OrderID","Product", "Quantity", "Date", "TotalSales", "Region"]].head()
```

### Aufgabe 3

Finde alle Verkäufe von Laptop oder Smartphone in Berlin oder Munich, bei denen der Rabatt (Discount) unter 10 % liegt und das Datum nach dem 1. März 2023 liegt.

```python
df.query("Product in ['Laptop', 'Smartphone'] and Store in ['Berlin', 'Munich'] and Discount < 0.10 and Date > '2023-03-01'")
```
