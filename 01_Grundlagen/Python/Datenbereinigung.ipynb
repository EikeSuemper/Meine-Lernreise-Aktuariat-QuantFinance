### 3.3 Deskriptive Statistik  

Nachdem ich die grundlegenden Informationen und die Struktur meines Datensatzes kenne, möchte ich nun die numerischen und kategorischen Merkmale genauer beschreiben. Die deskriptive Statistik hilft mir dabei, die Daten zusammenzufassen und erste Muster, Auffälligkeiten oder typische Werte zu erkennen ohne Hypothesen zu testen oder Modelle zu bauen.

Ich unterscheide dabei zwei Hauptbereiche:

- Numerische Variablen: Lage (Mittelwert, Median etc.), Streuung (Std, Min/Max, Quartile)
- Kategorische/ Text-Variablen: Häufigkeiten, Modalwert, Anteilswerte

Wichtigste pandas-Methode: `df.describe()`
Es gibt automatisch eine schöne Zusammenfassung (getrennt für Zahlen und für Kategorien, wenn man `include='object'` oder `include='all'` setzt).

```python
df.describe(include = "all")
```

```python
# Betrachtet nur kategorische und Textspalten/ Numerische Spalten werden ignoriert/ .T transponiert die Tabelle
cat_summary = df.describe(include=['object', 'category']).T     

# Berechnet den Anteil fehlender Werte pro Spalte
cat_summary['Anteil fehlend (%)'] = (df.isna().mean() * 100).round(2)
display(cat_summary)
```
