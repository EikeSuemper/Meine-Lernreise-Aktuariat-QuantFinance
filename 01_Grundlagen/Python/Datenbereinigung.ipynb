## Duplikate erkennen und entfernen  

Duplikate sind eines der häufigsten Probleme in Datensätzen. Diese entstehen z. B. durch:

- Mehrfache Uploads/ doppelte Imports
- Technische Fehler beim Zusammenführen von Datenquellen
- Mehrfache Einträge durch denselben Nutzer (z. B. doppelte Registrierung)
- Kopierfehler

Es gibt zwei Hauptarten von Duplikaten:

1. Vollständige Duplikate: 2 oder mehrere Zeilen besitzen die exakt selben Einträge
2. Teilduplikate: Gleiche logische Inhalte, aber kleine Unterschiede (z. B. gleiche Kunde_ID, aber unterschiedliche Schreibweisen)

Ich gehe in diesem Abschnitt beide Fälle durch.

### Vollständige Duplikate

Um vollständige Duplikate (Zeilen sind identisch) zu identifizieren, verwende ich zunächst den Befehl `.duplicated().sum()`, um die Anzahl der Duplikate festzustellen. Sollten nun Fehler vorhanden sein, so greife ich auf `.drop_duplicates(keep = "first")` zurück, sodass ich jeweils das erste Duplikate behalte. In diesem Fall existieren keine Duplikate.

```python
df.duplicated().sum()
```

### Teilduplikate

Hier prüfe ich, ob es bestimmte "Schlüsselspalten" gibt mit denselben Einträgen. In diesem Fall ist die Spalte "Kunde_ID" eine solche Schlüsselspalte. Auch in diesem Fall existieren keine Duplikate. 

```python
df.duplicated(subset = "Kunde_ID").sum()
```

## Textbereinigung und Stringoperationen

Textdaten (Spalten vom Typ `object` oder `string`) sind in realen Datensätzen fast immer verschmutzt.  
Typische Probleme sind:

- überflüssige Leerzeichen (vorne, hinten, mehrfach innen)
- inkonsistente Groß- und Kleinschreibung
- Sonderzeichen, Tippfehler, ungültige Einträge
- unterschiedliche Formate (z. B. „München“ vs. „Muenchen“ vs. „München “)
- Informationen, die in einem Feld stecken (z. B. PLZ + Stadt in einer Spalte)

Ziel dieses Abschnitts:  
saubere, einheitliche und nutzbare Textspalten erzeugen – mit möglichst wenigen manuellen Eingriffen. Ich bearbeite hier vor allem: `name`, `PLZ`, `Stadt`, `Email`, `Kategorie_letzter_Kauf`, `Geschlecht`, `Treuekarte`, `Newsletter_Status`

Zu Erinnerung rufe ich hier nochmal 20 Beispielzeilen des Datensatzes auf.

```python
df.sample(20)
```

### Groß- und Kleinschreibung  

Zum Thema Groß- und Kleinschreibung sind vor allem in den Spalten "Newsletter_Status", "Treuekarte" und "Kategorie_letzter_Kauf" verschiedene Varianten zu finden. Ich möchte die Anfangsbuchstaben alle groß haben. Dazu verwende ich die Funktion `.str.title()` und überschreibe die jeweiligen Spalten des Dataframes. Äquivalent dazu kann man mit `.str.lower()` die ersten Buchstaben verkleinern. 

```python
df["Kategorie_letzter_Kauf"]  = df["Kategorie_letzter_Kauf"].str.title()
df["Treuekarte"]              = df["Treuekarte"].str.title()
df["Newsletter_Status"]       = df["Newsletter_Status"].str.title()
df.sample(15)
```

### Sonderzeichen und ungültige Einträge entfernen oder ersetzen

Als nächstes möchte ich in der Spalte "Email" das "AT" durch "@" ersetzen. Dazu verwende ich `.str.replace("AT", "@")`.
Zudem möchte ich in der Spalte "PLZ" die Zeichen "-" und "D" entfernen.

```python
df["Email"] = df["Email"].str.replace("AT", "@")
df["PLZ"] = df["PLZ"].astype(str).str.replace(r"\D", "", regex=True)

df.sample(20)
```
