### Das statistische Modell  

Sei $\mathcal{X}$ die Menge von Beobachtungen (sogenannter Stichprobenraum). Ein messbarer Raum $(\mathcal{X}, \mathcal{F})$ zusammen mit einer Familie  
von Wahrscheinlichkeitsmaßen $(P_\theta)_{\theta \in \Theta}$ heißt statistisches Modell, kurz:

$$
(\mathcal{X}, \mathcal{F}, (P_\theta)_{\theta \in \Theta})
$$

ist ein statistisches Modell, wenn $(\mathcal{X}, \mathcal{F})$ ein messbarer Raum ist und $P_\theta$ für jedes $\theta \in \Theta$ ein Wahrscheinlichkeitsmaß auf $(\mathcal{X}, \mathcal{F})$ ist.
Eine Beobachtung oder Statistik ist eine messbare Abbildung

$$
X : \Omega \to \mathcal{X}
$$

die $(\mathcal{A}, \mathcal{F})$-messbar ist.

Diese Abbildung induziert ein neues statistisches Modell

$$
(\mathcal{X}, \mathcal{F}, (P_\theta^X)_{\theta \in \Theta})
$$

wobei $P_\theta^X = P_\theta \circ X^{-1}$ das Bildmaß ist. $X_1, \dots, X_n$ heißt eine Stichprobe,  
wenn $X_1, \dots, X_n : \Omega \to \mathcal{X}$ messbare Abbildungen sind. (üblicherweise iid unter jedem $P_\theta$).

## 1. Parameterschätzung

Wir betrachten ein parametrisches statistisches Modell

$$
(\mathcal{X}, \mathcal{F}, (P_\theta)_{\theta \in \Theta})
$$

mit  
- $\mathcal{X}$ = Beobachtungsraum  
- $\mathcal{F}$ = $\sigma$-Algebra  
- $\Theta$ = Parameterraum (meist Teilmenge von $\mathbb{R}^k$)  
- $P_\theta$ = Verteilung der Daten, abhängig vom unbekannten Parameter $\theta$

## 1.1 Grundbegriffe

### 1.1.1 Schätzer

Gegeben sei ein interessierender/ abgeleiteter Parameter (oder Parameterfunktion) 
$g : \Theta \to \mathbb{R}^p$. Ein Schätzer ist eine messbare Abbildung

$$
\hat{g} : \mathcal{X} \to \mathbb{R}^p, \quad x \mapsto \hat{g}(x)
$$

$\hat{g}(X)$ ist dann die zufällige Schätzung. $\hat{g}$ heißt unverzerrt oder erwartungstreu für $g(\theta)$, falls  
   $$
   \mathbb{E}_\theta[\hat{g}(X)] = g(\theta) \quad \text{für alle } \theta \in \Theta
   $$

### 1.1.2 Verlustfunktion

Eine Verlustfunktion ist eine messbare Abbildung

$$
L : \Theta \times \mathbb{R}^p \to [0, \infty)
$$

Das Risiko dieses Schätzers bei Parameter $\theta$ ist definiert als

$$
R(\theta, \hat{g}) = \mathbb{E}_\theta \bigl[ L(\hat{g}, g(\theta)) \bigr]
$$

Das ist die erwartete Verlustgröße beim Parameter $\theta$, wenn man den Schätzer $\hat{g}$ benutzt.


- **Quadratischer Verlust** 
   $$
   L(a, g(\theta)) = (a - g(\theta))^2 \quad \text{(für Skalarparameter)}
   $$

   $$
   L(a, g(\theta)) = \|a - g(\theta)\|^2 \quad \text{(für Vektorparameter)}
   $$

   → Risiko wird dann zum mittleren quadratischen Fehler (MSE):

   $$
   R(\theta, \hat{g}) = \mathbb{E}_\theta \bigl[ (\hat{g} - g(\theta))^2 \bigr] = \mathrm{Var}_\theta(\hat{g}) + \bigl( \mathbb{E}_\theta[\hat{g}] - g(\theta) \bigr)^2
   $$

Den Term
   $$
   \text{Bias}_\theta(\hat{g}) = \mathbb{E}_\theta[\hat{g}(X)] - g(\theta)
   $$
nennt man Bias.


- **Absoluter Verlust** 
   $$
   L(a, g(\theta)) = |a - g(\theta)|
   $$

   → Risiko = mittlerer absoluter Fehler (MAE)

- **0-1-Verlust** 
   $$
   L(a, g(\theta)) = 
   \begin{cases} 
   0 & \text{wenn } |a - g(\theta)| \leq \varepsilon \\
   1 & \text{sonst}
   \end{cases}
   $$
