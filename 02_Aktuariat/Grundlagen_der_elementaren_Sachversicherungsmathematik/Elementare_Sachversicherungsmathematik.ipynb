# Grundlagen der Sachversicherungsmathematik

## Modellierung
**Basis-Situation**

- Portfolio: $C := \{C_1, \dots, C_m\}$ Menge von Versicherungsverträgen.
- Gegenstand: Verluste des Portfolios in 1 Periode (z.B. 1 Bilanzjahr).
- Aggregierte Verluste: Nichtnegative ganzzahlige ZV $S$ mit Verteilung $P_S$ und VF $F_S$.
- $S$: Gesamtschaden-ZV.
- Basisrisiko: Ruin = $\{S > \pi\}$ für fixierte Prämie $\pi$. Nötig: Modelle für $S$, um Risiko zu schätzen.

### 1.1 Individuelles Modell

**Typische Schäden**

- $Z_1, \dots, Z_m$: Nichtnegative ganzzahlige ZV-en für Schäden aus $C_1, \dots, C_m$.
- $(Z_1, \dots, Z_m)$: Individuelles Modell für $S = \sum_{i=1}^m Z_i$, falls unabhängig.
- Homogenes Portfolio:$Z_i$ i.i.d. ~ $Z$ (typischer Schaden).

Dieses Modell summiert individuelle Risiken, um den Gesamtschaden zu modellieren. In homogenen Portfolios (viele ähnliche Verträge) nutzt man den "Ausgleich im Kollektiv": Die Gesamtprämie $\pi = m \cdot p$ (m = Anzahl Verträge, p = Prämie pro Vertrag) sorgt für Diversifikation. Bei großem m sinkt die Ruinwahrscheinlichkeit P(S > m p), solange p > E[Z] (Sicherheitszuschlag):

**Organisationsprinzip:** $\pi = m \cdot p$, Ausgleich im Kollektiv → kleine $P(S > m p)$.

Für homogenes Portfolio:
$$\lim_{m \to \infty} P(S > m p) = 
\begin{cases} 
1 & p < E[Z] \\ 
1/2 & p = E[Z] \\ 
0 & p > E[Z] 
\end{cases}$$
$p - E[Z] > 0$: Sicherheitszuschlag.

**Markov-Ungleichung**
$$P(|W| \geq \epsilon) \leq \frac{E[h(|W|)]}{h(\epsilon)}, \quad h \text{ monoton wachsend}.$$
Einfache Obergrenze für Abweichungswahrscheinlichkeiten, z.B. h(x)=x für P(W ≥ ε) ≤ E[W]/ε. Nützlich für Ruin-Schätzungen ohne volle Verteilung.

**Cantelli-Ungleichung**
$$P(W \geq E[W] + \epsilon) \leq \frac{\mathrm{Var}[W]}{\mathrm{Var}[W] + \epsilon^2}.$$
Einseitige Version von Chebyshev; schätzt positives Abweichen (z.B. hoher Schaden) unter Berücksichtigung der Varianz. Konservativ, aber varianzbasiert.
