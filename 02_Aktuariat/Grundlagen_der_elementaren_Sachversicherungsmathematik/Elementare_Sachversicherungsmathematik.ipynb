# Grundlagen der Sachversicherungsmathematik

## 1. Modellierung
**Basis-Situation**

- Portfolio: $C := \{C_1, \dots, C_m\}$ Menge von Versicherungsverträgen.
- Gegenstand: Verluste des Portfolios in 1 Periode (z.B. 1 Bilanzjahr).
- Aggregierte Verluste: Nichtnegative ganzzahlige ZV $S$ mit Verteilung $P_S$ und VF $F_S$.
- $S$: Gesamtschaden-ZV.
- Basisrisiko: Ruin = $\{S > \pi\}$ für fixierte Prämie $\pi$. Nötig: Modelle für $S$, um Risiko zu schätzen.

### 1.1 Individuelles Modell

**Typische Schäden/ Definition**

- $Z_1, \dots, Z_m$: Nichtnegative ganzzahlige ZV-en für Schäden aus $C_1, \dots, C_m$.
- $(Z_1, \dots, Z_m)$: Individuelles Modell für $S = \sum_{i=1}^m Z_i$, falls unabhängig.
- Homogenes Portfolio:$Z_i$ i.i.d. ~ $Z$ (typischer Schaden).

Dieses Modell summiert individuelle Risiken, um den Gesamtschaden zu modellieren. In homogenen Portfolios (viele ähnliche Verträge) nutzt man den "Ausgleich im Kollektiv": Die Gesamtprämie $\pi = m \cdot p$ (m = Anzahl Verträge, p = Prämie pro Vertrag) sorgt für Diversifikation. Bei großem m sinkt die Ruinwahrscheinlichkeit P(S > m p), solange p > E[Z] (Sicherheitszuschlag):

**Organisationsprinzip:** $\pi = m \cdot p$, Ausgleich im Kollektiv → kleine $P(S > m p)$.

Für homogenes Portfolio:
$$\lim_{m \to \infty} P(S > m p) = 
\begin{cases} 
1 & p < E[Z] \\ 
1/2 & p = E[Z] \\ 
0 & p > E[Z] 
\end{cases}$$
$p - E[Z] > 0$: Sicherheitszuschlag.

**Markov-Ungleichung**
$$P(|W| \geq \epsilon) \leq \frac{E[h(|W|)]}{h(\epsilon)}, \quad h \text{ monoton wachsend}.$$
Einfache Obergrenze für Abweichungswahrscheinlichkeiten, z.B. h(x)=x für P(W ≥ ε) ≤ E[W]/ε. Nützlich für Ruin-Schätzungen ohne volle Verteilung.

**Cantelli-Ungleichung**
$$P(W \geq E[W] + \epsilon) \leq \frac{\mathrm{Var}[W]}{\mathrm{Var}[W] + \epsilon^2}.$$
Einseitige Version von Chebyshev; schätzt positives Abweichen (z.B. hoher Schaden) unter Berücksichtigung der Varianz. Konservativ, aber varianzbasiert.





### 1.2 Kollektives Modell

**Definition**
Sei $S$ die nichtnegative, ganzzahlige Zufallsvariable des Gesamtschadens eines Portfolios.
Ein kollektives Modell zur Verteilung von $S$ besteht aus einer nichtnegativen, ganzzahligen Zufallsvariable $N$ (Schadenzahl) und einer Folge $(Xⱼ)ⱼ∈ℕ$ von nichtnegativen, ganzzahligen Zufallsvariablen (Schadenhöhen)

mit den Eigenschaften:

- $Xⱼ ~ X₁$ für alle $j ∈ ℕ$
- $X₁, …, Xₙ$ sind unabhängig für jedes $n ∈ ℕ$
- $N$ und $(X₁, …, Xₙ)$ sind unabhängig für jedes $n ∈ ℕ$
- $S = ∑_{j=1}^N Xⱼ $  (konventionell mit der Vereinbarung $∑_{j=1}^0 = 0)$

$N$ heißt Schadenzahl, $Xⱼ$ heißt Schadenhöhe zum j-ten Schaden.
Jede Zufallsvariable mit derselben Verteilung wie $X₁$ wird typische Schadenhöhe genannt. Die Verteilung lässt sich darstellen durch

$$
p_S(x) = P(N \in \{0\} \cap \{x\}) + \sum_{n=1}^\infty P(N=n) \cdot P\left( \sum_{j=1}^n X_j = x \right) \quad \text{für } x \in \mathbb{R}
$$


### 1.3 Panjer-Rekursion

Sei $(N, (X_j)_{j\in\mathbb{N}})$ ein **kollektives Modell** zur Verteilung des Gesamtschadens $S$ eines Portfolios mit typischer Schadenhöhe $X$ mit folgender Zähldichte:

$$
p_S(x) = P(N \in \{0\} \cap \{x\}) + \sum_{n=1}^\infty P(N=n) \cdot P\left( \sum_{j=1}^n X_j = x \right) \quad \text{für } x \in \mathbb{R}
$$

**Basisidee:**  
Wähle eine Verteilung für die Schadenzahl $N$, deren Zähldichte $p_N$ ein aus der **Panjer-Klasse** stammt.  
Diese rekursive Struktur überträgt sich dann auf die Zähldichte $p_S$ des Gesamtschadens → **Panjer-Rekursion**. Die Panjer-Klasse besteht aus allen diskreten Verteilungen auf $\mathbb{R}$ mit Zähldichte $q$ von folgender Gestalt:

Es existieren $a, b \in \mathbb{R}$ mit $a + b > 0$, so dass für alle $x \in \mathbb{R}$ gilt

$$
q(x) = 
\begin{cases} 
\left( a + \dfrac{b}{x} \right) q(x-1) & x \in \mathbb{N} \\
0 & \text{sonst}
\end{cases}
$$

$a$ und $b$ werden erste bzw. zweite Rekursionscharakteristik genannt. Die Klasse besteht aus exakt 3 Verteilungen:

| Verteilung                        | $a$                  | $b$                        |
|-----------------------------------|----------------------|----------------------------|
| $\mathrm{B}(n; p)$, $0 < p < 1$   | $-\dfrac{p}{1-p}$    | $\dfrac{(n+1)p}{1-p}$      |
| $\mathrm{Poi}(\lambda)$, $\lambda > 0$ | $0$                 | $\lambda$                  |
| $\mathrm{NB}(\alpha; p)$, $\alpha > 0$, $0 < p < 1$ | $\dfrac{p}{1-p}$   | $\dfrac{p(\alpha-1)}{1-p}$ |


Sei die Verteilung von $N$ aus der Panjer-Klasse mit Rekursionscharakteristiken $a, b$.

**Behauptung** (unter der Bedingung $a < 1$) gilt:

$$
P(S=0) = \sum_{n=0}^\infty P(N=n) \bigl[ P(X=0) \bigr]^n
$$

und für $k \in \mathbb{N}$

$$
P(S=k) = \dfrac{1}{1 - a \cdot P(X=0)} \sum_{l=1}^k \left( a + \dfrac{b}{k} \right) P(X=l) \, P(S = k-l)
$$


### 1.4 Das Ruin-Problem (Rest-Ruin-Risiko im individuellen und kollektiven Modell für homogene Portfolios)

Für ein **homogenes Portfolio** mit typischem Schaden $Z$ (Var$[Z] < \infty$) gelten Folgende Ruinwahrscheinlichkeiten:

$$
\lim_{m\to\infty} P(S > m \cdot p) = 
\begin{cases}
1   & \text{, falls } p < \mathbb{E}[Z] \\
\frac{1}{2} & \text{, falls } p = \mathbb{E}[Z] \\
0   & \text{, falls } p > \mathbb{E}[Z]
\end{cases}
$$

**Interpretation:**

- $p = \mathbb{E}[Z]$ → reine Nettoprämie (kein Sicherheitszuschlag) → Ruinwahrscheinlichkeit konvergiert gegen 50 %  
- $p > \mathbb{E}[Z]$ → positiver Sicherheitszuschlag → Ruinwahrscheinlichkeit → 0 (Diversifikationseffekt)  
- $p < \mathbb{E}[Z]$ → Unterdeckung (negativer Sicherheitszuschlag) → Ruin fast sicher

Dies begründet in dem späteren Kapitel "Prämienprinzipien" die Forderung eines nichtnegativen Siccherheitszuschlag.



## 2. Großschäden

### 2.1 Heavy-Tail-Verteilungen

**Warum sind Großschäden so wichtig?**

In der elementaren Sachversicherungsmathematik (insbesondere Naturkatastrophen, Haftpflicht, Industrie) dominieren oft wenige sehr große Schäden („Fat Tails“/ "Schwere Tails") den Gesamtschaden und damit das Ruinrisiko. Klassische Verteilungen wie Exponential- oder Gammaverteilungen haben leichte Schwänze, welche das Risiko extremer Ereignisse unterschätzen.

**Definition: Heavy-Tail-Verteilungen mit dem von Mises Kriterium**

Eine Verteilung zu einer stetigen Zufallsvariable hat einen **schweren Schwanz** (heavy tail), wenn die Tailwahrscheinlichkeit $1-F(x)$ langsamer abfällt als jede Exponentialverteilung, d. h. wenn für jedes $\varepsilon > 0$ gilt:

$$
\lim_{x\to\infty} \frac{1-F(x)}{e^{-\varepsilon x}} = +\infty
$$

Das ist das klassische "von Mises-Kriterium" für Verteilungen mit schwerem Schwanz.

**Leichter Schwanz** (light tail) liegt vor, wenn es ein $\varepsilon > 0$ gibt, sodass

$$
\limsup_{x\to\infty} \frac{1-F(x)}{e^{-\varepsilon x}} < \infty
$$

→ typisch für Exponential-, Gamma-, Normalverteilungen.

Ein schwerer Schwanz liegt vor, wenn die obige Grenzwertbeziehung für jedes $\varepsilon > 0$ erfüllt ist.

**Typische Heavy-Tail-Kandidaten in der Versicherungsmathematik**

| Verteilung     | Dichte / Verteilungsfunktion                          | Schwanzverhalten                  | Subexponentiell? | Momente existieren?          |
|----------------|--------------------------------------------------------|------------------------------------|------------------|-------------------------------|
| Pareto (Typ I) | $f(x) = \alpha x_m^\alpha x^{-(\alpha+1)}$, $x \ge x_m$ | $1-F(x) \sim x^{-\alpha}$         | ja               | $E[X^k]$ für $k < \alpha$     |
| Pareto (Typ II) / Lomax | $f(x) = \alpha \lambda^\alpha (\lambda + x)^{-(\alpha+1)}$ | $1-F(x) \sim x^{-\alpha}$         | ja               | $E[X^k]$ für $k < \alpha$     |
| Log-Normal     | $f(x) = \frac{1}{x \sigma \sqrt{2\pi}} \exp\left(-\frac{(\ln x - \mu)^2}{2\sigma^2}\right)$ | subexponentiell, aber sehr langsam | ja               | alle Momente existieren       |
| Burr (Typ XII) | $f(x) = c k \left(\frac{x}{\lambda}\right)^{c-1} \left(1 + \left(\frac{x}{\lambda}\right)^c \right)^{-(k+1)}$ | $1-F(x) \sim x^{-c k}$            | ja               | $E[X^r]$ für $r < c k$        |
| Weibull (mit $k<1$) | $f(x) = \frac{k}{\lambda} \left(\frac{x}{\lambda}\right)^{k-1} \exp\left(-\left(\frac{x}{\lambda}\right)^k\right)$ | $1-F(x) \sim \exp(-x^k)$          | nein (bei $k<1$) | alle Momente existieren       |

### 2.2 Momenterzeugende Funktionen und Verteilungen mit schwerem Tail

Die momenterzeugende Funktion $M_X(t)$ einer Zufallsvariablen $X$ ist definiert als

$$
M_X(t) = \mathbb{E}\bigl[e^{tX}\bigr]
$$

sofern dieser Erwartungswert endlich ist. Verteilungen mit schwerem Tail (heavy tail) sind dadurch charakterisiert, dass die momenterzeugende Funktion $M_X(t)$ für $t > 0$ nicht existiert (d. h. $\mathbb{E}[e^{tX}] = \infty$ für jedes $t > 0$).



Eine Verteilung hat leichten Schwanz, wenn es ein $\varepsilon > 0$ gibt, sodass

$$
\limsup_{x\to\infty} \frac{1-F(x)}{e^{-\varepsilon x}} < \infty
$$

Eine Verteilung hat einen schweren Tail, genau dann wenn

$$
\mathbb{E}[e^{tX}] = \infty \quad \text{für jedes } t > 0
$$

Das ist die definierende Eigenschaft einer heavy-tailed Verteilung.

- Leichter Tail ⇔ ∃ t > 0 mit M_X(t) < ∞  
- Schwerer Tail ⇔ ∀ t > 0 gilt M_X(t) = ∞

Dieses Kriterium ist äquivalent zur Aussage von Mises:

- Wenn der Schwanz langsamer abfällt als jede Exponentialfunktion → divergiert das Integral für jede positive exponentielle Gewichtung  
- → M_X(t) = ∞ für jedes t > 0

#### Beispiele

**Beispiel** Exponentialverteilung  
- M_X(t) existiert für t < λ (bei Rate λ)  
- → **leichter Tail**

**Beispiel** Pareto-Verteilung (Typ I)  
- M_X(t) existiert für t ≤ 0  
- → **schwerer Tail**

**Beispiel** Log-Normal-Verteilung  
- M_X(t) existiert für t ≤ 0  
- → **schwerer Tail** (obwohl alle Potenzmomente E[X^k] endlich sind)
 
### 2.3 Große Gesamtschäden im kollektiven Modell

Im kollektiven Modell $S = \sum_{j=1}^N X_j$ (N: Schadenzahl, X_j: i.i.d. Schadenhöhen) ist die Tail-Wahrscheinlichkeit $P(S > x)$ für große $x$ entscheidend. Bei heavy-tailed X: Asymptotisches Verhalten unterscheidet sich von leichten Tails.

#### 2.3.1 Asymptotisches Tail-Verhalten

Bei Verteilungen mit schwerem Tail gilt:

$$
P(S > x) \sim \mathbb{E}[N] \cdot P(X > x) \quad \text{für } x \to \infty
$$

**Bedeutung:**  
- Tail von $S$ proportional zum Tail von $X$, skaliert mit $\mathbb{E}[N]$.  
- Dominanz durch größten Einzel-Schaden – Summe kleiner Schäden vernachlässigbar.  
- Konsequenz: Ruin $P(S > \pi) \approx \mathbb{E}[N] \cdot P(X > \pi)$ – unabhängig von Varianz.


### 3. Prämienkalkulation

Prämienkalkulation zielt darauf ab, eine ausreichende Prämie für den Gesamtschaden $  S = \sum_{i=1}^m Z_i  $ eines Versicherungsportfolios zu bestimmen, um das Risiko des Versicherers zu managen. Ein Top-Down-Ansatz wird oft verwendet und erfolgt in dieser Reihenfolge:

- Festsetzung einer Gesamtprämie $  \rho(S)  $.
- Aufteilung (Allokation) auf individuelle Verträge.

#### 3.1 Prämienprinzipien

Prämienprinzipien sind Abbildungen $  \rho: L^\rho \to \mathbb{R}  $ mit $  L^\rho \subseteq L^1  $ (Menge absicherbarer Risiken, wo $  L^1  $ Risiken mit existierendem Erwartungswert sind), die folgende Eigenschaften erfüllen:

- Verteilungsinvarianz: $  \rho(V) = \rho(W)  $ wenn $  V \sim W  $. (2 Zufallsvariablen mit den selben Risiken sollen gleich hohe Prämien besitzen.)
- Mindestprämie: $  \rho(W) \geq E[W]  $ (Nettoprämie plus Sicherheitszulage).

Es gibt verschiedene Prämienprinzipien:

- **Nettoprämie-Prinzip:** $  \rho(W) = E[W]  $. Dies ist der "faire" Preis, führt aber langfristig zu Ruin, da Schwankungen nicht berücksichtigt werden.

- **Erwartungswert-Prinzip (Expected Value Principle):** $  \rho(W) = (1 + \rho) E[W]  $, mit Sicherheitsloading $  \rho > 0  $. Basierend auf dem starken Gesetz der großen Zahlen.

- **Varianz-Prinzip (Variance Principle):** $  \rho(W) = E[W] + \alpha \operatorname{Var}(W)  $, mit $  \alpha > 0  $.
Standardabweichungs-Prinzip (Standard Deviation Principle): $  \rho(W) = E[W] + \gamma \sqrt{\operatorname{Var}(W)}  $, mit $  \gamma > 0  $. Motiviert durch das zentrale Grenzwertsatz.

- **Hölländisches Prämienprinzip:** $  \rho(W) = E[W] + \gamma E[(W - E[W])_+]  $, mit $  \gamma > 0  $ und $  x_+ = \max\{x, 0\}  $.

- **Value at Risk (VaR):** $  \operatorname{VaR}_\alpha(W)  $, das $  \alpha  $-Quantil von $  W  $. Nicht immer ein Prämienprinzip, da es $  \rho(W) \geq E[W]  $ nicht immer erfüllt, aber das Perzentil-Prinzip auf geeigneten Mengen schon.

- **Average Value at Risk (AVaR) / Swiss Solvency:** $  \operatorname{AVaR}_\alpha(W) = \frac{1}{1-\alpha} \int_\alpha^1 \operatorname{VaR}_\beta(W) \, d\beta  $. Dies ist ein Prämienprinzip und monoton bezüglich stochastischer Ordnungen.

- **Wang-Prämienprinzip:** Basierend auf einer konkaven Verzerrungsfunktion $  g: [0,1] \to [0,1]  $, $  \rho_g(W) = \int_0^\infty g(\bar{F}_W(x)) \, dx  $, mit $  \bar{F}_W(x) = 1 - F_W(x)  $. Beispiele: Nettoprämie für $  g(t) = t  $, AVaR für spezifische $  g  $.


#### 3.2 Prämienallokation

Bei der Prämienallokation geht es nun um die Aufteilung der Gesamtprämie $\rho(S)$ auf die einzelnen Risiken $Z_1, Z_2, \dots, Z_m$  
im Portfolio, sodass jedes Risiko einen „fairen“ Beitrag zur Gesamtprämie leistet. Es gibt 2 Arten von Allokationen:

- **Konservative Allokation**  
  $\displaystyle \sum_{i=1}^m \pi_i \geq \rho(S)$  
Hierbei wird mindestens die Gesamtprämie erhoben, aber ein Sicherheitszuschlag ist möglich.

- **Volle Allokation** (meist gewünscht)  
  $\displaystyle \sum_{i=1}^m \pi_i = \rho(S)$  
Man erhebt die exakte Verteilung der gesamten Prämie ohne Überschuss oder Defizit.


Wichtige Beispiele für die Prämienallokation sind die Proportionale Allokation, die Kovarianz Allokation und ein Optimierungsproblem:

##### 3.2.1 Proportionale Allokation

Die einfachste und am häufigsten verwendete Methode ist die Proportionale Allokation:

$$
\pi_i = \frac{\rho(S)}{\sum_{j=1}^m \rho(Z_j)} \cdot \rho(Z_i)
$$

Jeder Vertrag erhält einen Anteil an der Gesamtprämie, der genau proportional zu seiner eigenen (isolierten) Risikoprämie $\rho(Z_i)$ ist.

##### 3.2.2 Kovarianz Allokation

$$
\pi_i = \frac{\rho(S)}{\operatorname{Var}(S)} \cdot \operatorname{Cov}(Z_i, S)
$$

Jeder Vertrag $i$ zahlt einen Beitrag zur Gesamtprämie, der proportional zu seinem Beitrag zur Gesamtvarianz des Portfolios ist.
- Risiken, die stark mit dem Gesamtschaden $S$ korrelieren (hohe Kovarianz), erhalten einen höheren Prämienanteil.
- Systematische Risiken (nicht-diversifizierbar) werden stärker gewichtet als idiosynkratische (diversifizierbare) Risiken.
- Besonders sinnvoll bei heterogenen Portfolios, bei denen Diversifikationseffekte eine große Rolle spielen.

##### 3.2.3 Optimierungsansatz zur Allokation

Hierbei handelt es sich um die Minimierung eines gewichteten quadratischen Abweichungsrisikos unter der Nebenbedingung der vollen Allokation:

$$
\min_{\pi_1,\dots,\pi_m} \quad
\sum_{i=1}^m E\left[ W_i \cdot \frac{(Z_i - \pi_i)^2}{v_i} \right]
\qquad \text{s.t.} \qquad
\sum_{i=1}^m \pi_i = \rho(S)
$$
 
Man sucht die Allokation $\pi = (\pi_1, \dots, \pi_m)$, die den erwarteten gewichteten quadratischen Fehler zwischen dem tatsächlichen Schaden $Z_i$ und der zugeteilten Prämie $\pi_i$ minimiert. Die Gewichtung $W_i$ und die Normalisierung $v_i$ erlauben eine flexible Anpassung an die Bedeutung und das Risikoprofil jedes Vertrags.

Dabei sind

- $W_i$ … Gewichtungsfaktor für Vertrag $i$  
- $v_i$ … Normalisierungsfaktoren (Skalierung des Fehlers)  

Der Zielfunktional ist quadratisch in den $\pi_i$ → konvexes Optimierungsproblem  
Die Nebenbedingung ist linear → das Problem ist gut lösbar.

Die Standardmethode ist der Lagrange-Ansatz:

$$
\mathcal{L}(\boldsymbol{\pi}, \lambda) = \sum_{i=1}^m E\left[ W_i \frac{(Z_i - \pi_i)^2}{v_i} \right] + \lambda \left( \rho(S) - \sum_{i=1}^m \pi_i \right)
$$

Partielle Ableitung nach $\pi_i$:

$$
\frac{\partial \mathcal{L}}{\partial \pi_i} = -2 E\left[ W_i \frac{Z_i - \pi_i}{v_i} \right] - \lambda = 0
\qquad \Rightarrow \qquad
\pi_i = E[Z_i] - \frac{\lambda \, v_i}{2 W_i}
$$

Einsetzen in die Nebenbedingung $\sum \pi_i = \rho(S)$ ergibt eine lineare Gleichung für $\lambda$ → **explizite Lösung** möglich.
