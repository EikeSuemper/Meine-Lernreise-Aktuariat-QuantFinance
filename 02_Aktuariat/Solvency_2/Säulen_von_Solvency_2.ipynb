### Die drei Säulen von Solvency II: Eine Zusammenfassung basierend auf der Richtlinie 2009/138/EG

Solvency II (Richtlinie 2009/138/EG) strukturiert den regulatorischen Rahmen für Versicherungs- und Rückversicherungsunternehmen in drei Säulen, die ein risikobasiertes Aufsichtssystem bilden. Diese Säulen zielen darauf ab, die finanzielle Stabilität zu gewährleisten, Risiken transparent zu managen und den Schutz der Versicherungsnehmer zu stärken. Die Säulen werden implizit durch die Kapitelstruktur der Richtlinie abgebildet, insbesondere in Titel I Kapitel IV (Governance und Veröffentlichung) und Kapitel VI (Quantitative Vorschriften). Im Folgenden fasse ich jede Säule kurz zusammen, mit Verweisen auf relevante Artikel innerhalb der Richtlinie als Quellen.

| Säule | Kerninhalt | Relevante Vorschriften (Quellen) |
|-------|------------|---------------------------------|
| **Säule 1: Quantitative Anforderungen**<br>(Kapitalanforderungen und Bewertung) | Diese Säule definiert harte Kapitalvorgaben: Bewertung von Vermögen und Verbindlichkeiten zu Marktwerten, Berechnung technischer Rückstellungen, Eigenmittel (in Tiers eingeteilt) sowie die Solvenzkapitalanforderung (SCR) und Mindestkapitalanforderung (MCR). Die SCR wird entweder per Standardformel (Risikomodule für Markt-, Lebens-, Nichtlebens- und operationelle Risiken) oder internem Modell (99,5%-Konfidenzniveau über 1 Jahr) ermittelt und deckt wesentliche Risiken ab. Ziel: Sicherstellung ausreichender Eigenmittel zur Risikobedeckung. | - Bewertung: Art. 75<br>- Rückstellungen: Art. 76–86<br>- Eigenmittel: Art. 87–99<br>- SCR: Art. 100–127 (Standardformel: Art. 103–111; Interne Modelle: Art. 112–127)<br>- MCR: Art. 128–131<br>- Anlagen: Art. 132–135<br>(Anhang IV: Standardformel-Details) |
| **Säule 2: Qualitative Anforderungen**<br>(Aufsichtliche Überprüfung und Governance) | Fokus auf internes Risikomanagement und Governance: Unternehmen müssen ein umfassendes Governance-System (inkl. Verwaltungsorgan, interne Kontrollen) einrichten. Zentral ist das Risikomanagementsystem (Identifikation, Messung, Überwachung von Risiken) und die eigene Risiko- und Solvenzbeurteilung (ORSA). Die Aufsichtsbehörden prüfen dies und können Kapitalaufschläge verhängen. Ziel: Förderung einer risikobewussten Unternehmenskultur. | - Allgemeine Governance: Art. 41–50<br>- Risikomanagement: Art. 44 (Leitlinien für Risiken wie Markt-, Kredit-, operationelle)<br>- ORSA: Art. 45<br>- Interne Kontrollen: Art. 46–47<br>(Kapitel III: Gruppenaufsicht mit Risikomanagement, Art. 246) |
| **Säule 3: Offenlegung und Transparenz**<br>(Berichterstattung und Markt-Discipline) | Regelt die regelmäßige Berichterstattung an Aufsichtsbehörden (z. B. SFCR – Solvency and Financial Condition Report) und öffentliche Offenlegung (z. B. zu Eigenmitteln, SCR, Risikoprofil). Ziel: Marktteilnehmer (z. B. Investoren) informieren, um Disziplin durch Marktkräfte zu schaffen. Jährliche Berichte und quantitative Templates sind vorgeschrieben. | - Veröffentlichung: Art. 51–56 (Öffentlicher Bericht, Inhalte zu SCR, Eigenmitteln)<br>- Berichterstattung: Art. 35, 129 (an Aufsichtsbehörden)<br>- Gruppenberichte: Art. 254 (SFR auf Gruppenebene) |

Diese Säulen interagieren: Säule 1 liefert die quantitativen Metriken, Säule 2 stellt sicher, dass Unternehmen sie intern handhaben, und Säule 3 sorgt für externe Transparenz.
